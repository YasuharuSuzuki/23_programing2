## サンプルプログラム2　棒グラフ
- 課題結果データで棒グラフを描画します。

#### 課題結果を可視化するクラスを作成する
- 個々のデータを格納するTestDataクラスと、全てのデータを管理するTestDataManagerクラスを作成する
- コンストラクタ（初期化関数）で課題の番号を指定するとファイル読み込みを行う
- ファイル読み込みはTestDataManager.readFile()関数で行う
- 可視化はTestDataManager.plot()関数で行う

```python
# 課題結果データクラスの定義
class TestData:
    # 初期化メソッド
    def __init__(self, kubun, count):
        self.kubun = kubun # 
        self.count = count

# 課題結果データManagerクラスの定義
class TestDataManager:
    # 初期化メソッド
    def __init__(self, number):
        self.number = number
        self.records = self.readFile(number)
    
    # ファイル読み込み
    def readFile(self, number):
        records = []
        with open(TEST_PATH.format(number), mode="r", encoding="utf_8") as f:
            all_lines = f.readlines()
            records = []
            for line1 in all_lines:
                record1 = line1.rstrip().split(',')
                records.append(TestData(record1[0],int(record1[1]))) # カンマ区切りで区分と人数が格納されているので、インスタンス化して取得
        return records

    # 全体の合計人数を取得
    def getTotalCount(self):
        return self.records[-2].count + self.records[-3].count
    
    # 可視化
    def plot(self):
        kubun = [record1.kubun for record1 in self.records]
        count = [record1.count for record1 in self.records]
        total_count = self.getTotalCount()
        fig, ax = plt.subplots(figsize=(15,8))# サイズを指定
        bar1 = ax.bar(kubun, count, label="test{:02}".format(self.number)) # X軸、Y軸両方、凡例を指定。
        ax.bar_label(bar1, label_type='edge', padding=0, color='blue') # 棒グラフにラベルを設定
        ax.set_xticks(range(len(kubun))) # tick location を指定。0〜表示ラベル数分のindexをtick locationとします。
        ax.set_xticklabels(kubun) # 表示ラベルを変更
        ax.xaxis.set_tick_params(rotation=0) # X軸の表示を90度回転させる。
        ax.yaxis.set_major_formatter('{x:.0f}人') # Y軸に単位をつける
        ax.legend(loc="upper left") # 凡例を有効にする
        ax.grid(axis='y') # グリッドを有効にする
        ax.set_title("とある課題{:02}の結果(提出数:{})".format(self.number, total_count)) # タイトルを表示。日本語が文字化けしてしまう場合はjapanize_matplotlibを使用してください。
```

#### 実際に棒グラフを描画してみる
- 課題の読み込みはTestDataManagerクラスを初期化するだけ
- 可視化はTestDataManager.plot()関数を呼び出すだけ

```python
import matplotlib.pyplot as plt
import japanize_matplotlib # matplotlibで日本語表示できない時にpipでライブラリをインストールして、有効にしてみてください

test_data_manager = TestDataManager(3)
test_data_manager.plot()
plt.show()
```

## サンプルプログラム4　積み上げ棒グラフ
- 産業別GDPデータで積み上げ棒グラフを描画します。

#### GDPデータを可視化するクラスの作成
- 個々のデータを格納するGDPDataクラスと、全てのデータを管理する GDPDataManager クラスを作成する
- コンストラクタ（初期化関数）で読み込むデータを指定するとファイル読み込みを行う
- 可視化は GDPDataManager.plot() 関数で行う

```python
# GDPデータクラスの定義
class GDPData:
    # 初期化メソッド
    def __init__(self, year, items, total):
        self.year = year
        self.items = list(map(float, items))
        self.total = float(total)

# GDPデータManagerクラスの定義
class GDPDataManager:
    # 初期化メソッド
    def __init__(self, industry_type, gdp_type):
        self.readFile(industry_type, gdp_type)

    # タイトルのパース
    def __parseCaption(self, all_lines):
        self.caption = all_lines[13].rstrip().split(',')[1].strip('"').replace('【','').replace('】','')
#         self.caption = all_lines[0].rstrip().split(',')[1].strip('"')
#         print(self.caption)
        
    # 列名のパース
    def __parseColumnNames(self, all_lines):
        self.column_names = [column.strip('"') for column in all_lines[5].rstrip().split(',')[1:10]]
#         print(self.column_names)

    # 各年のデータのパース
    def __parseYealyData(self, all_lines):
        records = []
        for year in range(14,len(all_lines)):
            rows = [value.strip('"') for value in all_lines[year].rstrip().split(',')]
            data = GDPData(rows[0], rows[1:10], rows[11])
            records.append(data)
#             print(data.year, data.items, data.total)
        self.records = records
        
    # ファイル読み込み
    def readFile(self, industry_type, gdp_type):
        all_lines = []
        with open(GDP_DATA_PATH.format(industry_type, gdp_type), mode="r", encoding="cp932") as f:
            all_lines = f.readlines()
        self.__parseCaption(all_lines)
        self.__parseColumnNames(all_lines)
        self.__parseYealyData(all_lines)

    # 可視化
    def plot(self):
        fig, ax = plt.subplots(figsize=(15,8))# サイズを指定
        years = [record.year for record in self.records]
        x_pos = range(len(years))
        
        color_map = plt.get_cmap("tab20b")
        column_sums = [0] * len(self.records)
        bars = []
        legend_labels = []
        for item_index in range(len(self.column_names)):
            column1 = self.column_names[item_index]
            column_years =  [record.items[item_index] for record in self.records]
            bar1 = ax.bar(x_pos, column_years, label=self.column_names[item_index], bottom=column_sums, # X軸、Y軸両方、凡例を指定。
                                color=color_map(item_index/len(self.column_names)))
            ax.bar_label(bar1, label_type='center', padding=0, color='white') # 棒グラフにラベルを設定
            bars.append(bar1)
            column_sums =  [column_sums[column_index]+column_years[column_index] for column_index in range(len(column_years))]
            legend_labels.append(self.column_names[item_index])

        ax.set_xticks(x_pos, years)    # X軸ラベル（垂直）
        ax.legend(bars, legend_labels, loc = "upper left")    # 凡例を作る
        ax.yaxis.set_major_formatter('{x:.0f}兆') # Y軸に単位をつける
        ax.grid(axis='y') # グリッドを有効にする
        ax.set_title(self.caption) # タイトルを表示。日本語が文字化けしてしまう場合はjapanize_matplotlibを使用してください。            
```

#### 実際に積み上げ棒グラフを描画してみる
- GDPデータの読み込みは GDPDataManager クラス を初期化するだけ
- 可視化は GDPDataManager.plot() 関数を呼び出すだけ

```python
industry_nominal = GDPDataManager('industry','nominal')
industry_nominal.plot()
```

## サンプルプログラム5　散布図
- Pythonで有名なあやめのデータで散布図を描画します。
<img src="./image/petal_sepal_label.png">

#### あやめのデータを可視化するクラスの作成
- 個々のデータを格納するIrisDataクラスと、全てのデータを管理する IrisDataManager クラスを作成する
- コンストラクタ（初期化関数）で読み込むデータを指定するとファイル読み込みを行う
- 可視化は IrisDataManager.plot() 関数で行う

```python
# あやめデータクラスの定義
class IrisData:
    # 初期化メソッド
    def __init__(self, SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm, Species):
        self.SepalLengthCm = float(SepalLengthCm)
        self.SepalWidthCm = float(SepalWidthCm)
        self.PetalLengthCm = float(PetalLengthCm)
        self.PetalWidthCm = float(PetalWidthCm)
        self.Species = Species

# あやめデータManagerクラスの定義
class IrisDataManager:
    # 初期化メソッド
    def __init__(self):
        self.readFile()

    # 各あやめのデータのパース
    def __parseData(self, all_lines):
        records = []
        for index in range(1,len(all_lines)):
            rows = [value.strip('"') for value in all_lines[index].rstrip().split(',')]
            data = IrisData(rows[1], rows[2], rows[3], rows[4], rows[5])
            records.append(data)
        self.records = records

    # 種ごとの各あやめのデータの作成
    def __parseSpecies(self):
        self.species = list(set([record.Species for record in self.records])) # 種の種類をsetで取得→listで取得
        self.species_dict = dict()
        for species1 in self.species:
            self.species_dict[species1] = list(filter(lambda x: x.Species == species1, self.records))
        
    # ファイル読み込み
    def readFile(self):
        all_lines = []
        with open(IRIS_PATH, mode="r", encoding="utf-8") as f:
            all_lines = f.readlines()
        self.__parseData(all_lines)
        self.__parseSpecies()

    # 可視化
    def plot(self, x_label, y_label):
        fig, ax = plt.subplots(figsize=(15,8))# サイズを指定
        scatters = []
        for species1 in self.species:
            x_values = [getattr(record, x_label) for record in self.species_dict[species1]]
            y_values = [getattr(record, y_label) for record in self.species_dict[species1]]
            scatters.append(ax.scatter(x_values, y_values, label=species1))    # グラフを描く

        ax.legend(scatters, self.species, loc = "upper left")    # 凡例を作る
        ax.xaxis.set_major_formatter('{x:.1f} cm') # X軸に単位をつける
        ax.yaxis.set_major_formatter('{x:.1f} cm') # Y軸に単位をつける
        ax.grid() # グリッドを有効にする
        ax.set_xlabel(x_label)
        ax.set_ylabel(y_label)
        ax.set_title("あやめのデータ({}/{})".format(x_label, y_label)) # タイトルを表示。日本語が文字化けしてしまう場合はjapanize_matplotlibを使用してください。            
```

#### 実際にあやめの散布図を描画してみる
- GDPデータの読み込みは IrisDataManager クラス を初期化するだけ
- 可視化は IrisDataManager.plot() 関数を呼び出すだけ

```python
iris_manager = IrisDataManager()
iris_manager.plot("SepalLengthCm", "SepalWidthCm")
```

- コンストラクタ（初期化関数）で読み込むデータを指定するとファイル読み込みを行う
- 可視化は InfoGDPDataManager.plot() 関数で行う

```python
# InfoGDPDataデータクラスの定義
class InfoGDPData:
    # 初期化メソッド
    def __init__(self, year, items, total):
        self.year = year
        self.items = list(map(float, items))
        self.total = float(total)

# InfoGDPデータManagerクラスの定義
class InfoGDPDataManager:
    # 初期化メソッド
    def __init__(self, industry_type, gdp_type):
        self.readFile(industry_type, gdp_type)

    # タイトルのパース
    def __parseCaption(self, all_lines):
        self.caption = all_lines[13].rstrip().split(',')[1].strip('"').replace('【','').replace('】','')
#         print(self.caption)
        
    # 列名のパース
    def __parseColumnNames(self, all_lines):
        self.column_names = [column.strip('"') for column in all_lines[5].rstrip().split(',')[1:10]]
#         print(self.column_names)

    # 各年のデータのパース
    def __parseYealyData(self, all_lines):
        records = []
        for year in range(14,len(all_lines)):
            rows = [value.strip('"').replace('－','0.0') for value in all_lines[year].rstrip().split(',')]
            data = InfoGDPData(rows[0], rows[1:9], rows[10])
            records.append(data)
#             print(data.year, data.items, data.total)
        self.records = records
        
    # ファイル読み込み
    def readFile(self, industry_type, gdp_type):
        all_lines = []
        with open(GDP_DATA_PATH.format(industry_type, gdp_type), mode="r", encoding="cp932") as f:
            all_lines = f.readlines()
        self.__parseCaption(all_lines)
        self.__parseColumnNames(all_lines)
        self.__parseYealyData(all_lines)

    # 可視化
    def plot(self, year):
        fig, ax = plt.subplots(figsize=(15,8))# サイズを指定
        items = [record.items for record in filter(lambda x: x.year == year, self.records)][0]
        color_map = plt.get_cmap("tab20b")
        colors = [color_map(index/len(items)) for index in range(len(items))]
        column_names = self.column_names[0:len(items)]

        patches, texts, autotexts = ax.pie(items, labels=column_names, colors=colors, autopct='%1.1f%%', startangle=90, counterclock=False, shadow=True)
        ax.legend(column_names, loc="upper left")    # 凡例を作る
        for autotext in autotexts:
            autotext.set_color('white')

        ax.set_title("{}({}年)".format(self.caption, year)) # タイトルを表示。日本語が文字化けしてしまう場合はjapanize_matplotlibを使用してください。            
```

#### 実際に情報産業国内生産の円グラフを描画してみる
- GDPデータの読み込みは InfoGDPDataManager クラス を初期化するだけ
- 可視化は InfoGDPDataManager.plot() 関数を呼び出すだけ

- 課題データ管理クラス TestDataManager() を継承して作成します

```python
# 課題結果データManagerクラスの定義
class TestDataManagerSubplot(TestDataManager):
    # 可視化
    def subplot(self, ax, x_label_rotation=0):
        kubun = [record1.kubun for record1 in self.records]
        count = [record1.count for record1 in self.records]
        total_count = self.getTotalCount()
        bar1 = ax.bar(kubun, count, label="test{:02}".format(self.number)) # X軸、Y軸両方、凡例を指定。
        ax.bar_label(bar1, label_type='edge', padding=0, color='blue') # 棒グラフにラベルを設定
        ax.set_xticks(range(len(kubun))) # tick location を指定。0〜表示ラベル数分のindexをtick locationとします。
        ax.set_xticklabels(kubun) # 表示ラベルを変更
        ax.xaxis.set_tick_params(rotation=x_label_rotation) # X軸の表示を90度回転させる。
        ax.yaxis.set_major_formatter('{x:.0f}人') # Y軸に単位をつける
        ax.legend(loc="upper left") # 凡例を有効にする
        ax.grid(axis='y') # グリッドを有効にする
        ax.set_title("とある課題{:02}の結果(提出数:{})".format(self.number, total_count)) # タイトルを表示。日本語が文字化けしてしまう場合はjapanize_matplotlibを使用してください。
```
